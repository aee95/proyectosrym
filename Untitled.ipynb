```python
def transformadaHadamard(imagenGris):
    H1 = np.array([[1, 1], [1, -1]]) 
    #print(H1)
    HB = np.array([[1, 1], [1, -1]]) 
    #print(HB)
    tam = len(imagenGris)
    num = int(np.log2(tam))
    for i in range(num-1):
        Hn = np.kron(H1, HB)
        HB = Hn
    #print("H",num,"=",Hn)
    #print(len(Hn))
    HnT = np.transpose(Hn)
    F = np.dot(np.dot(Hn,imagenGris),HnT)
    print(F)
    return F
```

```python
imagentransformada = transformadaHadamard(imagenReferenciaGris)
```

```python
#np.seterr(divide = 'ignore') 
#transform = fwht(imagenReferenciaGris) 
#print(transform)
img_hadamard = np.log(np.abs(imagentransformada)) 
print(img_hadamard)
plt.imshow(img_hadamard, cmap = 'gray')
plt.title('Magnitude Spectrum Hadamard'), plt.xticks([]), plt.yticks([])
plt.show()
```
